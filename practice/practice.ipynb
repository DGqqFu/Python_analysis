```python
#PM2.5 < 10 的站點

#PM2.5 前 10 糟的站點
import csv

cs = []

with open('aqx_p_488.csv', encoding='utf-8', newline='') as file_a:
    reader = csv.DictReader(file_a)
    
    for a in reader:
        v = a['pm2.5']
        if v.strip() != '':
            pm = float(v)
            if pm < 10:
                sn = a['sitename']
                cs.append((sn, pm))

print("PM2.5 < 10 的站點有:")
for s in cs:
    print(f"{s[0]} 的 PM2.5 是 {s[1]}")




            
    
```

```python
print("PM2.5 前 10 糟的站點為:")
m = sorted(cs, key=lambda x: x[1], reverse=True)[:10]

for s in m:
    print(f"{s[0]} 的 PM2.5 是 {s[1]}")
```
