```python
import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib.font_manager import fontManager
import random

fontManager.addfont("ChineseFont.ttf")
mpl.rc('font', family="ChineseFont")

days=list(range(1,31))
july=[round(random.uniform(28.5,35.2),ndigits=1 )for _ in range(1,31)]
august=[round(random.uniform(28.5,35.2),ndigits=1)for _ in range(1,31)]
print(july)
print(august)
```

```python
'''plt.figure(figsize=(8,12))
plt.plot(days,july,marker='o',linestyle='-.',label="七月",alpha=0.5)
plt.plot(days,august,marker='o',linestyle='-.',label="八月",alpha=0.5)
plt.legend()
plt.grid(axis='y')
plt.title('7月溫度')'''

plt.figure(figsize=(8,12))
plt.subplot(2, 1, 1)  # 2列1欄的第1張圖
plt.plot(days, july, marker='o', linestyle='-.', label="七月", color='red', alpha=0.5)
plt.title('7月溫度')
plt.legend()
plt.grid(axis='y')

# 第2張子圖：八月
plt.subplot(2, 1, 2)  # 2列1欄的第2張圖
plt.plot(days, august, marker='o', linestyle='-.', label="八月", color='blue', alpha=0.5)
plt.title('8月溫度')
plt.legend()
plt.grid(axis='y')

plt.show()
```

```python
figure = plt.figure(figsize=(8, 12))
axes1 = figure.add_subplot(2,1,1)
axes1.plot(days, july,marker='o',color='r',linestyle='-.',label='7月')
axes1.set_xticks(days)
axes1.legend()
axes1.grid(axis='y')
axes1.set_title('7月溫度')
axes1.set_xlabel("日期")
axes1.set_ylabel("攝氏")

axes2 = figure.add_subplot(2,1,2)
axes2.scatter(days, august,marker='x',label='八月',color='g')
axes2.set_xticks(days)
axes2.legend()
axes2.grid(axis='y')
axes2.set_title('8月溫度')
axes2.set_xlabel("日期")
axes2.set_ylabel("攝氏")
plt.tight_layout()
plt.show()
plt.close()
```
